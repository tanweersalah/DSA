```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
```

```python
import time
import matplotlib.pyplot as plt
from random import randint

# Function to measure execution time of a given function
def measure_time_complexity( max_input_size=1000):
    input_sizes = range(0, max_input_size + 1)
    times = []

    mylist = []

    for size in input_sizes:
        input_data = range(size)
        index = randint(0,size)

        start_time = time.time()

        # function
        binary_search(input_data, index)
        # end
        
        end_time = time.time()
        times.append(end_time - start_time)

    return input_sizes, times
```

```python
input_sizes, times = measure_time_complexity( max_input_size=1000)
```

```python
plt.plot(input_sizes, times)
plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Time Complexity of Different Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```
