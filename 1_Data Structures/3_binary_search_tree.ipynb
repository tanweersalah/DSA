A Binary Search Tree is a data structure used in computer science for organizing and storing data in a sorted manner. Each node in a Binary Search Tree has at most two children, a left child and a right child, with the left child containing values less than the parent node and the right child containing values greater than the parent node. This hierarchical structure allows for efficient searching, insertion, and deletion operations on the data stored in the tree.

### Methods of Binary Search Tree (BST)

1. **Insert**
   - **Description**: Adds a new node with a given value to the BST, maintaining the BST property.
   - **Time Complexity**: 
     - Average Case: \( O(\log n) \)
     - Worst Case: \( O(n) \)

2. **Delete**
   - **Description**: Removes a node with a given value from the BST, maintaining the BST property.
   - **Time Complexity**:
     - Average Case: \( O(\log n) \)
     - Worst Case: \( O(n) \)

3. **Search**
   - **Description**: Finds a node with a given value in the BST.
   - **Time Complexity**:
     - Average Case: \( O(\log n) \)
     - Worst Case: \( O(n) \)

4. **Find Minimum**
   - **Description**: Finds the node with the smallest value in the BST.
   - **Time Complexity**: 
     - Average and Worst Case: \( O(\log n) \) or \( O(n) \) (depending on the tree's balance)

5. **Find Maximum**
   - **Description**: Finds the node with the largest value in the BST.
   - **Time Complexity**:
     - Average and Worst Case: \( O(\log n) \) or \( O(n) \) (depending on the tree's balance)

6. **Inorder Traversal**
   - **Description**: Visits all the nodes in ascending order (left-root-right).
   - **Time Complexity**: \( O(n) \)

7. **Preorder Traversal**
   - **Description**: Visits all the nodes in the order (root-left-right).
   - **Time Complexity**: \( O(n) \)

8. **Postorder Traversal**
   - **Description**: Visits all the nodes in the order (left-right-root).
   - **Time Complexity**: \( O(n) \)

9. **Height**
   - **Description**: Calculates the height of the BST, which is the length of the longest path from the root to a leaf.
   - **Time Complexity**: \( O(n) \)

10. **Find Successor**
    - **Description**: Finds the node with the smallest value greater than a given node.
    - **Time Complexity**: 
      - Average Case: \( O(\log n) \)
      - Worst Case: \( O(n) \)

11. **Find Predecessor**
    - **Description**: Finds the node with the largest value smaller than a given node.
    - **Time Complexity**:
      - Average Case: \( O(\log n) \)
      - Worst Case: \( O(n) \)

### Summary of Big O Notations

- **Average Case**: Operations like insert, delete, search, find minimum, find maximum, find successor, and find predecessor typically have \( O(\log n) \) time complexity due to the balanced nature of the tree.
- **Worst Case**: For unbalanced BSTs (which can degenerate into a linked list), these operations have \( O(n) \) time complexity.
- **Traversal and Height Calculation**: Both have \( O(n) \) time complexity as they involve visiting every node in the tree.

### PLot

```python
import matplotlib.pyplot as plt

def calculate_positions(node, depth=0, pos_x=0, pos=None, level_spacing=1, sibling_spacing=1):
    if pos is None:
        pos = {}
    if node:
        pos[node] = (pos_x, -depth * level_spacing)
        calculate_positions(node.leftChild, depth + 1, pos_x - sibling_spacing, pos, level_spacing, sibling_spacing / 2)
        calculate_positions(node.rightChild, depth + 1, pos_x + sibling_spacing, pos, level_spacing, sibling_spacing / 2)
    return pos



def plot_tree(node, pos=None):
    if pos is None:
        pos = calculate_positions(node)
    
    fig, ax = plt.subplots()
    for node, (x, y) in pos.items():
        ax.scatter(x, y, s=100)
        ax.text(x, y, node.value, ha='center', va='center', color='white', fontsize=10, bbox=dict(facecolor='blue', edgecolor='black', boxstyle='circle,pad=0.3'))

        if node.leftChild:
            lx, ly = pos[node.leftChild]
            ax.plot([x, lx], [y, ly], 'k-')

        if node.rightChild:
            rx, ry = pos[node.rightChild]
            ax.plot([x, rx], [y, ry], 'k-')
    
    ax.axis('off')
    plt.show()
```

## BST

```python
class BinarySearchTree():
    def __init__(self) -> None:
        self.head = None


    def insert(self,value):
        node = Node(value)
        if self.head :
            current_node = self.head
            cont = True
            while cont:
                if value < current_node.value  :
                    if current_node.leftChild :
                        current_node = current_node.leftChild
                    else :
                        current_node.leftChild = node
                        cont = False    
                else :
                    if current_node.rightChild :
                        current_node = current_node.rightChild
                    else :
                        current_node.rightChild = node
                        cont = False  
        else :
            node = Node(value)
            self.head = node
    
    def search(self,value):
        current_node = self.head
        
        while current_node :
            if current_node.value == value :
                return current_node
            elif value < current_node.value  :
                if current_node.leftChild :
                    current_node = current_node.leftChild
                else :
                    current_node = None
            else  :
                if current_node.rightChild :
                    current_node = current_node.rightChild
                else :
                    current_node = None
        return current_node
    
    def find_max(self):
        current_node = self.head
        max_value = current_node
        while current_node:
            max_value = current_node
            current_node= current_node.rightChild
        return max_value

    def find_min(self):
        current_node = self.head 
        min_value = current_node
        while current_node:
            min_value = current_node
            current_node = current_node.leftChild
        return min_value
    


    def in_order_traversal(self):
        value_list = []

        self._in_order_transversal( self.head,value_list)

        return value_list

    def _in_order_transversal(self, node,result):
        if node :
            self._in_order_transversal( node.leftChild,result)
            result.append(node.value)
            self._in_order_transversal( node.rightChild,result)






```
